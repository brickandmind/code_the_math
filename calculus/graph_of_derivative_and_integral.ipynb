```python
import matplotlib.pyplot as plt
import seaborn as sns
import math
import pandas as pd
```

## is derivative of sin(x) really cos(x)?

```python
lis = []
for x in range(100):
    sin_val = math.sin(x)
    cos_val = math.cos(x)
    dsin = (math.sin(x + 0.001) - math.sin(x))/0.001
    lis.append({'sin':sin_val, 'cos':cos_val, 'dsin':dsin,'x': x})
    
```

```python
df= pd.DataFrame(lis)
fig, ax = plt.subplots(nrows = 1, figsize = (20, 7))
plt.plot(df['x'], df['sin'], label = 'sin')
plt.plot(df['x'], df['cos'], label = 'cos')
plt.scatter(df['x'], df['dsin'], label = 'dsin')
plt.legend()
plt.show()

# plt.plot(x = df['x'], y= df['cos'])
```

## is derivative of $x^3$ is really $2x^2$?

```python
lis = []
for x in range(10):
    dx = ((x+0.0000000001)**3 - x**3)/0.0000000001
    dx_func = (2*x)**2
    lis.append({'dx':dx, 'dx_func':dx_func, 'x': x})
    
```

```python
df= pd.DataFrame(lis)
fig, ax = plt.subplots(nrows = 1, figsize = (20, 7))
plt.plot(df['x'], df['dx'], label = 'dx')
plt.scatter(df['x'], df['dx_func'], label = 'dx_func')
plt.legend()
plt.show()

# plt.plot(x = df['x'], y= df['cos'])
```
